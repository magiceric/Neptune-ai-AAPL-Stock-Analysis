```python
import logging
import time
import sys
from tqdm import tqdm
from utils.StockHistoryFetcher import StockHistoryFetcher

# Setup logging
log_filename = f"logs/history_fetcher_{time.strftime('%Y%m%d')}.log"
logging.basicConfig(filename=log_filename, level=logging.INFO, format='%(asctime)s - %(levelname)s - %(message)s', encoding='utf-8')

# Redirect stderr to log file
original_stderr = sys.stderr  # Save the original stderr
log_file = open(log_filename, 'a', encoding='utf-8')

# Create an instance of StockHistoryFetcher
fetcher = StockHistoryFetcher(None)

try:
    
    # Get all stock codes
    # Redirect stdout to log file for fetcher outputs
    original_stdout = sys.stdout  # Save the original stdout
    sys.stdout = log_file
    all_stock_codes_df = fetcher.fetch_all_stock_codes()
    # Loop through all stock codes and fetch historical data for each with a progress bar
    # sys.stdout = original_stdout  # Restore stdout to console after fetching

    for index, row in tqdm(all_stock_codes_df.iterrows(), total=all_stock_codes_df.shape[0], desc="Fetching History", file=sys.stdout):
        stock_code = row['code']
        fetcher.stock_code = stock_code  # Update the stock code in fetcher
        logging.info(f"Fetching history for stock code: {stock_code}")
        
        # sys.stdout = log_file  # Redirect stdout to log file
        fetcher.fetch_history()
        sys.stdout = original_stdout  # Restore stdout to console after fetching
finally:
    # Ensure resources are released after fetching is done
    logging.info("Completed fetching all stock histories.")
    del fetcher

# Restore stderr
log_file.close()

logging.shutdown()
```

```python
import time
import sys
import pandas as pd
from tqdm.notebook import tqdm  # Changed from tqdm to tqdm.notebook for Jupyter Notebook compatibility
from sqlalchemy import create_engine, text
from StockAnalysisTool import StockAnalysisTool

# Setup for console output
global_timestamp = time.strftime("%Y-%m-%d_%H-%M-%S")

# Setup for console output redirection
log_filename = f"logs/M_V_B_K_signal_hunter_{time.strftime('%Y%m%d')}.log"
# sys.stdout = open(log_filename, 'a')
# sys.stderr = open(log_filename, 'a')

mining_date = '2024-05-22'
specified_date = '2024-02-23'

# Connect to the database
engine = create_engine('mysql+pymysql://stock:Abcd1234!!@192.168.3.17:3306/aistock')
query = f"SELECT id, stockCode FROM goldencross WHERE miningDate = '{mining_date}'"
stock_codes = pd.read_sql(query, engine)

# Clear the mvbkSignal and mvbkHuntingDate fields for records with the current miningDate
clear_query = text("UPDATE goldencross SET mvbkSignal = NULL, mvbkHuntingDate = NULL, jAtHuntingDate = NULL WHERE miningDate = :mining_date")
with engine.connect() as connection:
    connection.execute(clear_query, {'mining_date': mining_date})
    connection.commit()

try:
    # Ensure tqdm gets the total count by passing the length of the DataFrame
    for index, row in tqdm(stock_codes.iterrows(), total=stock_codes.shape[0], desc="Analyzing Stocks", file=sys.stdout):  # Changed file=sys.stdout to ensure tqdm output and other prints go to the log file
        stock_code = row['stockCode']
        stock_id = row['id']
        # Remove the '.SZ' or '.SH' suffix from the stock code
        formatted_stock_code = stock_code.rstrip('.SZ').rstrip('.SH')
        tool = StockAnalysisTool(formatted_stock_code, global_timestamp)  # Create an instance of StockAnalysisTool for each stock code
        
        # Redirecting print statements to log file
        original_stdout = sys.stdout  # Save a reference to the original standard output
        with open(log_filename, 'a') as f:
            # sys.stdout = f  # Change the standard output to the file we created.
            print(f"Analyzing stock code: {formatted_stock_code}")

            # Check conditions for the stock
            days_back = 7
            moving_average_conditions, ma5_slope = tool.check_moving_average_conditions(specified_date, days_back)
            moving_average_conditions = 1 if moving_average_conditions else 0
            diff_above_dea, last_gc_before_hunting_date = tool.check_diff_above_dea(specified_date, days_back)
            diff_above_dea = 1 if diff_above_dea else 0
            price_above_bollinger = 1 if tool.check_price_above_bollinger_middle(specified_date, days_back) else 0
            kdj_crossover = 1 if tool.find_recent_kdj_condition(specified_date, 3) else 0
            volume_ma_crossover = 1 if tool.check_moving_average_conditions(specified_date, 10) else 0  # New condition
            price_up_volume_increase = 1 if tool.check_price_up_and_volume_increase(specified_date) else 0  # New condition
            # Count the number of conditions met
            conditions_met = sum([moving_average_conditions, diff_above_dea, price_above_bollinger, kdj_crossover, volume_ma_crossover, price_up_volume_increase])

            if conditions_met == 6:
                print(f"Stock {formatted_stock_code} meets all criteria on {specified_date}")
                update_query = text("UPDATE goldencross SET mvbkSignal = true, mvbkHuntingDate = :specified_date, ma5slope = :ma_slope, lastGCbeforeHuntingDate = :last_gc_before_hunting_date WHERE id = :stock_id")
                with engine.connect() as connection:                                            
                    connection.execute(update_query, {'stock_id': stock_id, 'specified_date': specified_date, 'ma_slope': ma5_slope, 'last_gc_before_hunting_date': last_gc_before_hunting_date})
                    connection.commit()
        # sys.stdout = original_stdout  # Reset the standard output to its original value
        del tool  # Release the tool instance after processing each stock

finally:
    # Release resources
    engine.dispose()
    # sys.stdout.close()
    # sys.stderr.close()
```
